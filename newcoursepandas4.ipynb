```python
import seaborn as sns
import pandas as pd
```

```python
tt=sns.load_dataset("titanic")
tt.head()
```

```python
tt.groupby("sex")[["survived"]].sum()
```

```python
tt.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
tt.pivot_table("survived",index="sex",columns="class")
```

```python
ageyas=pd.cut(tt["age"],[2,18,90])
ageyas.head()
```

```python
tt.pivot_table("survived",["sex",ageyas],"class")
```
