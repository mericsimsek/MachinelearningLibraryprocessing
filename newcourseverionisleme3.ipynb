```python
import numpy as np
import pandas as pd
V1=np.array([1,3,6,np.NaN,12,32,24,np.NaN,3])
V2=np.array([3,4,2,12,np.NaN,12,2,24,np.NaN])
V3=np.array([4,1,12,2,42,3,21,np.NaN,1])
df=pd.DataFrame({"V1":V1,"V2":V2,"V3":V3}) #unutma böyle bir işlemi yani liste halinde yazmayı ("v1":V1i ) süslü parantezle yaparsın
df
```

```python
df.dropna(how="all") #tüm nan olan satırları sil
```

```python
df.dropna(axis=1) #sütunda nan varsa sil
```

```python
df.dropna(axis=1,how ="all")#sütunun hepsi nan ise sil
```

```python
df["V1"].fillna(df["V1"].mean())
```

```python
df.apply(lambda x: x.fillna(df["V2"].mean()),axis=0)
```

```python
df.fillna(df.mean()[:])
```

```python
df.fillna(df.mean())
```

```python
df.fillna(df.mean()["V1":"V2"])
```

```python
df["V3"].fillna(df.median()["V3"])
```

```python
df["V3"].fillna(df["V3"].median())# ikinci yol medyana atamak  ilki ortalamayı atamaktı
```

```python
df.where(pd.notna(df),df.mean(),axis="columns")#eksik değerlerin yerine ort atama
```

```python
#kategorik ort. atama yani arge departmanındaki ort maaşlarla  pazarlamadaki ort. maşşlarlar aynı lmayacağından
#arge çalışanının eksik maaşını arge ort maaşınkısiyle atanır  pazarlamadaki ayrı atanır
V1=np.array([1,3,6,np.NaN,12,32,24,np.NaN,3])
V2=np.array([3,4,2,12,np.NaN,12,2,24,np.NaN])
V3=np.array([4,1,12,2,42,3,21,np.NaN,1])
V4=np.array(["IT","IT","Ik","IT","Ik","IT","Ik","Ik","IT"])
df=pd.DataFrame({"maas":V1,"V2":V2,"V3":V3,"departman":V4}) #unutma böyle bir işlemi yani liste halinde yazmayı ("v1":V1i ) süslü parantezle yaparsın
df.groupby("departman")["maas"].mean()
```

```python
df["maas"]
```

```python
df["maas"].fillna(df.groupby("departman")["maas"].transform("mean"))
```
