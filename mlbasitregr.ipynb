```python
#modellemeye artÄ±k baÅŸlÄ±yoruz ve ben seÃ§im olarak energy seÃ§ecem adam tv seÃ§miÅŸ 
import statsmodels.api as sm
X=df[["energy"]]
X.head()
```

```python
X=sm.add_constant(X)#matris iÅŸlemi iÃ§in gerekiyor
X.head() #bu baÄŸÄ±msÄ±z deÄŸiÅŸkendir ÅŸimdi baÄŸÄ±mlÄ± deÄŸiÅŸken alacaÄŸÄ±z
```

```python
#baÄŸÄ±mlÄ± deÄŸiÅŸken olarak loudness i aldÄ±m
y=df ["loudness"]
lm=sm.OLS(y,X)#model kurma iÅŸlemi gerÃ§ekleÅŸti
```

```python
modelmeric=lm.fit()
modelmeric.summary() 
#r kare deÄŸiÅŸkeni(0.517) Ã§ok Ã¶nemli ÅŸu baÄŸÄ±msÄ±z deÄŸiÅŸkenin baÄŸÄ±mlÄ± deÄŸiÅŸkeni aÃ§Ä±klama baÅŸarÄ±sÄ±dÄ±r  aÃ§Ä±klamadÄ±r yani
#f statisctic var anlamlÄ±lÄ±ÄŸÄ±n anlaÅŸÄ±labilmesi iÃ§in   prob  p valuedir
```
