```python
#modellemeye artık başlıyoruz ve ben seçim olarak energy seçecem adam tv seçmiş 
import statsmodels.api as sm
X=df[["energy"]]
X.head()
```

```python
X=sm.add_constant(X)#matris işlemi için gerekiyor
X.head() #bu bağımsız değişkendir şimdi bağımlı değişken alacağız
```

```python
#bağımlı değişken olarak loudness i aldım
y=df ["loudness"]
lm=sm.OLS(y,X)#model kurma işlemi gerçekleşti
```

```python
modelmeric=lm.fit()
modelmeric.summary() 
#r kare değişkeni(0.517) çok önemli şu bağımsız değişkenin bağımlı değişkeni açıklama başarısıdır  açıklamadır yani
#f statisctic var anlamlılığın anlaşılabilmesi için   prob  p valuedir
```

```python
#en önemli yer 2.tablo    const yazı yazıdğımız b0 değeridir  coefte katsayısıdır(parametredir)   standart hatası  p value (p >t
#yani kat sayı anlamlı diypor  0.000 yani anlamlı  ve  yüzde 18bin761 değerlendirilmiş)
#sabit(b0)  y ekseninin kestiği noktadır diğer değişkenlerde hiç bir değişkenlik olmadığnıda y nin beklenen değeridir  energynin 
#bi etkissi olmadığında  ortalama loudness -13.16 dir b0 katsayısının (sabitin yorumu budur)
#energynin yorumu energyde 1 birimlik artış olduğunda laudnesste  9.38birimlik artış olr  y =loudness seçtim şiddet album
#anlamlılık doğru yerinde  , güven aralığı bu 0.025 ve 0.975 arasındaki yani 18bin761 dir
```
